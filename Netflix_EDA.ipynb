#  To know the major streaming categories in netflix

```python
# Display the count of each category in the 'type' column
print(df['type'].value_counts())
```

```python
# Plotting a pie chart for the 'type' column
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

# Spliting the Data(df) into two DataFrames


```python
movies = df[df['type'] == 'Movie']
```

```python
# Display the number of rows and columns in the 'movies' DataFrame
print("Number of Rows in Movies DataFrame:", movies.shape[0])
print("Number of Columns in Movies DataFrame:", movies.shape[1])
```

```python
tv_shows=df[df['type'] == 'TV Show']
```

```python
# Display the number of rows and columns in the 'tv_shows' DataFrame
print("Number of Rows in TV Shows DataFrame:", tv_shows.shape[0])
print("Number of Columns in TV Shows DataFrame:", tv_shows.shape[1])
```

# Insights  from TVshows

```python
# Check for missing values in the 'tvshows' DataFrame
print(tv_shows.isna().sum())
```

```python
# sample dataframe
tv_shows.head(3)
```

```python
# Display the count of each rating category in the 'tv_shows' DataFrame
print(tv_shows['rating'].value_counts())
```

```python
def tvshows_rating():
    tv_shows['rating'].value_counts().plot(kind='bar', color='skyblue')
    plt.title("Distribution of TV Show Ratings on Netflix")
    plt.xlabel("Rating")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.show()

# Call the function to generate the plot
tvshows_rating()
```

```python
def top_tv_countries(x):
    top_countries = tv_shows['country'].value_counts().head(x)
    
    plt.figure(figsize=(12, 6))
    top_countries.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    
    plt.title(f"Top {x} Countries Producing TV Shows on Netflix")
    plt.ylabel('')  # Remove y-axis label
    
    plt.show()

# Call the function to generate the plot for the top 10 countries
top_tv_countries(10)
```

```python
# Display the names of the columns in the 'tv_shows' DataFrame
print(tv_shows.columns)
```

```python
def top_tv_years(x):
    top_years = tv_shows['release_year'].value_counts().head(x)
    
    plt.figure(figsize=(12, 6))
    top_years.plot(kind='bar', color='lightcoral')
    
    plt.title(f"Top {x} Years with Most TV Shows Released on Netflix")
    plt.xlabel("Year")
    plt.ylabel("Count")
    
    plt.show()

# Call the function to generate the plot for the top 15 years
top_tv_years(10)
```

```python
def top_tv_genres(x):
    top_genres = tv_shows['listed_in'].value_counts().head(x)
    
    plt.figure(figsize=(12, 6))
    top_genres.plot(kind='bar', color='skyblue')
    
    plt.title(f"Top {x} TV Show Genres on Netflix")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right', fontsize=10)
    
    plt.show()

# Call the function to generate the bar plot for the top 15 genres
top_tv_genres(10)
```

```python
def top_tvshow_genres(x):
    plt.figure(figsize=(24, 4))  # Set the figure size before plotting
    tv_shows['listed_in'].value_counts().head(x).plot(kind='pie', autopct='%1.1f%%', startangle=90)
    
    plt.title(f"Top {x} TV Show Genres on Netflix")
    plt.ylabel('')  # Remove y-axis label
    
    plt.show()

# Call the function to generate the pie chart for the top 5 genres
top_tvshow_genres(5)
```

```python
def top_movie_prod(x):
    plt.figure(figsize=(24, 8))  # Set the figure size before plotting
    movies['country'].value_counts().head(x).plot(kind='pie', autopct='%1.1f%%', startangle=90)
    
    plt.title(f"Top {x} Countries Producing Movies on Netflix")
    plt.ylabel('')  # Remove y-axis label
    
    plt.show()

# Call the function to generate the pie chart for the top 10 countries
top_movie_prod(10)
```

```python
def top_movie_count(x):
    plt.figure(figsize=(12, 6))
    movies['country'].value_counts().head(x).plot(kind='bar', color='deepskyblue')
    
    plt.title(f"Top {x} Countries Producing Movies on Netflix")
    plt.xlabel("Country")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
    
    plt.show()

# Call the function to generate the bar plot for the top 10 countries
top_movie_count(10)
```

```python
def top_movie_releasedyear(x):
    plt.figure(figsize=(12, 6))
    movies['release_year'].value_counts().head(x).plot(kind='bar', color='lightcoral')
    
    plt.title(f"Top {x} Years with Most Movie Releases on Netflix")
    plt.xlabel("Year")
    plt.ylabel("Count")
    plt.xticks(rotation=0)  # Adjust rotation for better readability
    
    plt.show()

# Call the function to generate the bar plot for the top 15 years
top_movie_releasedyear(15)
```

```python
def top_movie_genere(x):
    plt.figure(figsize=(18, 10))  # Set the figure size before plotting
    movies['listed_in'].value_counts().head(x).plot(kind='bar', color='lightcoral')
    
    plt.title(f"Top {x} Movie Genres on Netflix")
    plt.xlabel("Movie Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
    
    plt.show()

# Call the function to generate the bar plot for the top 10 movie genres
top_movie_genere(10)
```

```python
def top_movie_datesadded(x):
    plt.figure(figsize=(12, 6))
    movies['date_added'].value_counts().head(x).plot(kind='bar', color='skyblue')
    
    plt.title(f"Top {x} Dates with Most Movies Added on Netflix")
    plt.xlabel("Date Added")
    plt.ylabel("Count of Movies added")
    plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
    
    plt.show()

# Call the function to generate the bar plot for the top 10 dates
top_movie_datesadded(10)
```

```python
# to find the total number of minutes ,movie content avilabe in netflix
def total_movie_minutes():
    movies['duration_min'] = pd.to_numeric(movies['duration_min'], errors='coerce')
    return movies['duration_min'].sum()

total_minutes = total_movie_minutes()
print("Total Minutes of content:", total_minutes)
```

```python
# top list of directors who did more films
def top_movie_director(x):
    return movies.director.value_counts().head(x)
top_movie_director(10)
```
